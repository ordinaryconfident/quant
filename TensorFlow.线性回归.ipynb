```python
import tensorflow as tf
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
a=np.linspace(1,50)
b=3*a+6+np.random.rand(1,50)*5
b=b.reshape(-1,1)
plt.scatter(a,b)
plt.show()
```

```python
b
```

```python
model=tf.keras.Sequential()
```

```python
model.add(tf.keras.layers.Dense(1,input_shape=(1,)))
```

```python
model.compile(optimizer='adam',
              loss='mse',
               
)
```

```python
history=model.fit(a,b,epochs=10000)
```

```python
model.predict(np.array([5,]))
```

```python
model.add(tf.keras.layers.Dense(10,input_shape=(1,),activation='relu'))
model.add(tf.keras.layers.Dense(5,input_shape=(10,),activation='relu'))
model.add(tf.keras.layers.Dense(1))
```

```python
model.summary()
```

```python
model.compile(optimizer='adam',
              loss='mse',
               
)
```

```python
history=model.fit(a,b,epochs=500)
```
