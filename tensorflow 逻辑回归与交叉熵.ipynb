```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#心跳
x1=np.random.normal(70,4,size=(200,1))
x2=np.random.normal(80,4,size=(200,1))
```

```python
#身高
x3=np.random.normal(169.5,5,size=(200,1))
x4=np.random.normal(158,5,size=(200,1))
```

```python
#体重
x5=np.random.normal(66.2,5,size=(200,1))
x6=np.random.normal(57.3 ,5,size=(200,1))
```

```python
#头发颜色
#黑色=0，金色=1，其他颜色=2
list_head=[0,1,2]
x7=np.random.choice(list_head,size=(200,1))
x8=np.random.choice(list_head,size=(200,1))
```

```python
#婚姻情况
#已婚=0,未婚=1
list_head=[0,1]
x9=np.random.choice(list_head,size=(200,1))
x10=np.random.choice(list_head,size=(200,1))
```

```python
x_male=np.hstack((x1,x3,x5,x7,x9))
x_female=np.hstack((x2,x4,x6,x8,x10))
```

```python
y_male=np.zeros([50,1])
y_female=np.ones([50,1])
y_train=np.vstack([y_male,y_female])
```

```python
x_train_male=x_male[:50,]
x_train_female=x_female[:50,]
x_train=np.vstack([x_train_male,x_train_female])
```

```python
model=tf.keras.Sequential()
```

```python
model.add(tf.keras.layers.Dense(10, input_shape=(5,),activation='relu'))
model.add(tf.keras.layers.Dense(10,activation='relu'))
model.add(tf.keras.layers.Dense(1,activation='sigmoid'))
```

```python
model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['acc']
)
```

```python
history=model.fit(x_train,y_train,epochs=500)
```

```python
x_test=x_male[51:100,]
y_predict=model.predict(x_test)
y_predict
```
