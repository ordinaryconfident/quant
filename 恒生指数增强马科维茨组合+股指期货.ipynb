```python
ticker_list=['0016.HK','2269.HK','1876.HK','0268.HK','1810.HK','2202.HK','0823.HK','0868.HK','0981.HK','1928.HK']
```

```python
def month_data(x,y):
    stock_data = {}
    for ticker in ticker_list: 
        stock=yf.download(ticker,start=f"2022-{x}-01",end=f'2022-{x}-{y}')
        stock_data[ticker] = stock
    price_data= pd.DataFrame()
    for i in ticker_list: 
         price=stock_data[i]['Adj Close']
         price_data[i]=price 
    stock_log_ret = np.log(price_data/price_data.shift(1))
    stock_price=stock_log_ret[1:]
    return stock_price
```

function for weight

```python
def random_weight(n, m):
    numbers = [random.random() for _ in range(n)]
    summation = sum(numbers)
    k = m / summation
    return [i * k for i in numbers]
```

```python
def get_best_weight(month):
    port_weights=[]
    port_sharpe=[]
    for i in range(1000):
        weights=np.array(random_weight(10,1))
        return_portfolio=np.dot(month.sum(),weights)-rf
        volatility =np.sqrt(np.dot(weights.T, np.dot(month.cov()*len(month), weights))) 
        sharpe= return_portfolio/volatility
        port_sharpe.append(sharpe)
        port_weights.append(weights)
    best_port_stock=port_weights[port_sharpe.index(sharpe.max())]
    return best_port_stock
```
